```python

# Colonnes contenant du JSON stringifié
json_columns = ["services", "horaires"]
# Nettoyer les champs JSON stringifiés -> en JSON brut (non échappé)
for col in json_columns:
    if col in df.columns:
        df[col] = df[col].apply(
            lambda x: json.dumps(json.loads(x), ensure_ascii=False) if isinstance(x, str) and pd.notna(x) else x
        )

# Export CSV si besoin
df.to_csv("stations.csv", sep=";", index=False, quoting=3, escapechar='\\')
```
